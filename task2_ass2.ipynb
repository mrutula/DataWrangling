** Performing the same changes to the company names as done in task 1 using fuzzy logic **

```python
con['Company'] = con['Company'].astype(str)
#removing the leading and trailing spaces from the company names
con['Company'] = con['Company'].map(lambda x: x.strip())
#removing the '.' at the end of each company name 
con['Company'] = con['Company'].map(lambda x: x.replace(".","") if re.search(r'[.]$',x) else x)
# replacing the double spaces with single space
con['Company'] = con['Company'].map(lambda x: re.sub(r"\s\s+"," ", x) if re.search(r'\s\s+',x) else x)
#converting the company names to lower case
con['Company'] = con['Company'].map(lambda x: x.lower())  
```

```python
# repalcing ltd with limited , serv. with services and nan with empty

con['Company'] = con['Company'].map(lambda x: re.sub(r"ltd","limited",x) if re.search(r'\b(ltd)$',x) else x)
con['Company'] = con['Company'].map(lambda x: re.sub(r"ltd","limited",x) if re.search(r'\b(ltd)\b',x) else x)
con['Company'] = con['Company'].map(lambda x: re.sub(r"ltd","limited",x) if re.search(r'(ltd)',x) else x)
con['Company'] = con['Company'].map(lambda x: re.sub(r"\bserv\.\b","services",x) if re.search(r'serv\.',x) else x)
con['Company'] = con['Company'].map(lambda x: x.replace("nan","") if re.search(r'\bnan\b',x) else x)
```

```python
con_com_uni = con.Company.unique()
con_com_uni = list(con_com_uni)
```

```python
# replacing company names having no spaces with its counterpart with space

for i in con_com_uni:
    if ' ' in i:
        i_nospace = ''.join(i.split())
        if i_nospace in con_com_uni:
            con.loc[con.Company == i_nospace, 'Company'] = i
            
```

```python
# replacing a company name with its longer counterpart
for i in con_com_uni:
    if ' ' in i:
        subseti = set(i.split())
        for j in con_com_uni:
            if ' ' in j:
                subsetj = set(j.split())
                if i != j and subseti.issubset(subsetj):
                    con.loc[con.Company == i, 'Company'] = j
```

```python
# * 'easywebrecruitment', 'easyweb recruitment'
for i in con_com_uni:
    if ' ' in i:
        k = i.replace(' ','')
        for j in con_com_uni:
            if i != j and k == j:
                con.loc[con.Company == j, 'Company'] = i
```

```python
# replacing the company names having two spaces with company names having only one 
for i in con_com_uni:
    if ' ' in i:
        words = i.split()
        k =words[0]+''+words[1] 
        for j in range(2,len(words)): 
            k = k + ' ' + words[j]
        for m in con_com_uni:
            if k == m:
                con.loc[con.Company == i, 'Company'] = m 
```

```python
# replacing the names without '&' with its counterpart with '&'
for i in con_com_uni:
    if '&' in i:
        k = i.replace('&',' ')
        subseti = set(k.split())
        for j in con_com_uni:
            if ' ' in j:
                subsetj = set(j.split())
                if i != j and subseti.issubset(subsetj):
                    con.loc[con.Company == j, 'Company'] = i
```

```python
# replacing names without \' with its counterpart having \'
for i in con_com_uni:
    temp = i
    if re.search(r'\'',temp):
        k = re.sub(r'\'',' ',temp)
        for j in con_com_uni:
            if i != j and k == j:
                con.loc[con.Company == j, 'Company'] = i
```
