### Some of the salary have ranges and others have a 'K' denoting 1000

```python
# checking those records having a range of salary values
job[job['Salary per annum'].str.contains('-')]
```

```python
# replacing the range with mean value
sal_uni = job['Salary per annum'].unique()
for i in sal_uni:
    if '-' in i:
        sal_split = i.split('-')
        sal_split = list(map(float, sal_split))
        avg_sal = sum(sal_split)/len(sal_split)
        job.loc[job['Salary per annum'] == i,'Salary per annum'] = avg_sal
```

```python
# replacing K with 1000
job['Salary per annum'] = (job['Salary per annum'].replace(r'[K]+$','',regex = True).astype(float)* job['Salary per annum'].str.extract(r'[\d\.]+([K]+)', expand=False).fillna(1).replace('K',10**3).astype(int))
```

** Since some of the salaries were in the form eg. 34K **
    * job.Salary_per_annum.replace(r'[K]+$','',regex = True).astype(float) strips the K
    * job.Salary_per_annum.str.extract(r'[\d\.]+([K]+)', expand=False).fillna(1).replace('K',10**3).astype(int) this converts the K into 1000 and multiplies to the original number

```python
#converting the Salary_per_annum column to integer
job['Salary per annum'] = job['Salary per annum'].astype(int)
```

```python
job_company_uni = list(job_company_uni)
```

```python
# replacing company names having no spaces with its counterpart with space

for i in job_company_uni:
    if ' ' in i:
        i_nospace = ''.join(i.split())
        if i_nospace in job_company_uni:
            job.loc[job.Company == i_nospace, 'Company'] = i
            
            
            
```

```python
# replacing a company name with its longer counterpart
for i in job_company_uni:
    if ' ' in i:
        subseti = set(i.split())
        for j in job_company_uni:
            if ' ' in j:
                subsetj = set(j.split())
                if i != j and subseti.issubset(subsetj):
                    job.loc[job.Company == i, 'Company'] = j
        
                
```

```python
# * 'easywebrecruitment', 'easyweb recruitment'
for i in job_company_uni:
    if ' ' in i:
        k = i.replace(' ','')
        for j in job_company_uni:
            if i != j and k == j:
                job.loc[job.Company == j, 'Company'] = i
```

```python
# replacing the company names having two spaces with company names having only one 
for i in job_company_uni:
    if ' ' in i:
        words = i.split()
        k =words[0]+''+words[1] 
        for j in range(2,len(words)): 
            k = k + ' ' + words[j]
        for m in job_company_uni:
            if k == m:
                job.loc[job.Company == i, 'Company'] = m 
                
            
```

```python
# replacing the names without '&' with its counterpart with '&'
for i in job_company_uni:
    if '&' in i:
        k = i.replace('&',' ')
        subseti = set(k.split())
        for j in job_company_uni:
            if ' ' in j:
                subsetj = set(j.split())
                if i != j and subseti.issubset(subsetj):
                    job.loc[job.Company == j, 'Company'] = i
                    
```

```python
# replacing names without \' with its counterpart having \'
for i in job_company_uni:
    temp = i
    if re.search(r'\'',temp):
        k = re.sub(r'\'',' ',temp)
        for j in job_company_uni:
            if i != j and k == j:
                job.loc[job.Company == j, 'Company'] = i
```
